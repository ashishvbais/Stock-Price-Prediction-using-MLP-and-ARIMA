### Importing libraries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r'/Users/avbais/Documents/data.csv',index_col=0) #Importing csv data
data.index = pd.to_datetime(data.index) #Converting date into index for ARIMA
from pyramid.arima import auto_arima
```

### X Axis for Graph Plotting

```python
x_axis = np.zeros((200,1))
for t in range(0,200):
    x_axis[t,0] = t 
```

### Setting variables

```python
#Setting variables
index = 1000 #Selecting first 1000 line items for each stock
acc_matrix = np.zeros((5,1)) # Setting Zero Matrix to store accuracy values from stock
```

### Loop for setting ARIMA and Calculating Error, Accuracy and Mean Forecast Value of Selected Stocks

```python
# Selecting first 5 stocks for analysis    
for i in range(0,5):
    
    print('###################################################################################################################')
    print('')
    print('stock number: ' + str(i+1))
    print('')
    
    #Selecting stocks from dataset
    stocks = data.iloc[0:index,i:i+1] # Used iLoc as it is dataframe to use numbers
    
    #Setting ARIMA 
    stepwise_model = auto_arima(stocks, start_p=1, start_q=1,max_p=3, max_q=3, m=12,start_P=0, seasonal=True,d=1, D=1, trace=True,error_action='ignore',suppress_warnings=True, stepwise=True)
    print(stepwise_model.aic())
    print('')
    
    #Allocating test and train data
    train = stocks.iloc[0:800]
    test = stocks.iloc[800:]
    
    #Training data
    stepwise_model.fit(train)
    
    #Predicting values for forecasting stock prices
    future_forecast = stepwise_model.predict(n_periods=300)
    
    test = test.values
    
    #Converting forecast values from array to 1d matrix for better code compatiability
    future_val = np.reshape(future_forecast,(300,1))
    
    # Prediction for first 200 values
    prediction = future_val[0:200,:]
    
    # Forecast Values
    forecast = future_val[200:300,:]
    
    # Calculating Error
    error = test - prediction
    error_sq = np.multiply(error,error)
    error_sq_sum = error_sq.sum()
    error_sq_mean = error_sq_sum / 200
    mse = np.sqrt(error_sq_mean) 
    
    print('Mean Square error for stock ' + str(i+1) + ' is: ' + str(mse))
    
    # Calculating Accuracy
    acc = 1 - abs(np.divide(error,test))
    acc_mean = acc.sum() / 200
    acc_percentage = (acc_mean)*100
    acc_vector = acc * 100
    print('Accuracy percentage for stock ' + str(i+1) + ' is: ' + str(acc_percentage))
    
    #Calculating Mean Forecast Value for Stock
    mean_forecast = forecast.sum() / 100
    print('Mean forecast value of future stock ' + str(i+1) + ' is: ' +str(mean_forecast))
    
    #Plotting one stock over 40K line items
    stocks_predicted = stocks
    stocks_val = stocks.values
    plt.plot(stocks_val)
    
    #Plotting predicted value for same stock
    stocks_predicted_val = stocks_predicted.values
    stocks_predicted_val[800:] = prediction
    plt.plot(stocks_predicted_val)
    plt.show()
    
    print('')
    print('')
    
    x_axis = np.reshape(x_axis,(200))
    x = x_axis
    acc_vector = np.reshape(acc_vector,(200))
    y = acc_vector
    z = np.polyfit(x, y, 3)
    func = np.poly1d(z)
    #Stock price accuracy with respect time function 
    print('Function of Accuracy with respect to time, for this stock is: ' + str(np.poly1d(func)))

    acc_matrix[i,0] = mean_forecast
```

### Identifying best performing stock based on mean forecast value of stock 

```python
#Identifying best performing stock based on Mean Forecast value of stock   
for j in range(0,5):
    if acc_matrix[j,0] == acc_matrix.max():
        print('Best performing stock in future is: ' + 'Stock number '+ str(j+1))
```
